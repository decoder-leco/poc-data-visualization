```typescript
import pl from "npm:nodejs-polars";
import { display } from "https://deno.land/x/display@v0.1.1/mod.ts";
```

```typescript

let df_atih = pl.DataFrame({ 
        "foo": [1,2,3],
        "bar": [6.0, 7.0, 8.0],
        "ham": ["a", "b", "c"],
    }
)

let df_RD = pl.DataFrame(
    {
        "apple": ["x", "y", "z"],
        "ham": ["a", "b", "d"],
    }
)

const bidule = df_atih.join(df_RD, {on: "ham"})
await display(bidule);
//console.log(bidule)
```

```typescript
let response = await fetch( "https://raw.githubusercontent.com/decoderleco/deces_europe/main/data/csv/Atih.csv", );

let data_atih = await response.text();
data_atih = data_atih.replace(/"/g, "")
console.log("data_atih: ", data_atih.substring(0,50))

response =  await fetch( "https://raw.githubusercontent.com/decoderleco/deces_europe/main/data/csv/departements-region.csv", );
let data_DR = await response.text();
data_DR = data_DR.replace(/num_dep/, "dep");
console.log("data_DR: ", data_DR.substring(0,50))

let df_atih = pl.readCSV(data_atih, { sep: "," });
let df_DR = pl.readCSV(data_DR, { sep: "," });

const bidule = df_atih.join(df_DR, {on: "dep"}).tail(2)

await display(bidule);
```

```typescript
import * as Plot from "npm:@observablehq/plot";
import { DOMParser, SVGElement } from "npm:linkedom";
const document = new DOMParser().parseFromString(
  `<!DOCTYPE html><html lang="en"></html>`,
  "text/html",
);

let records = bidule.tail(1).toRecords();
console.log(records[0])
```

```typescript
function randomColor() {
    const R = Math.floor(Math.random()*99)
    const G = Math.floor(Math.random()*99)
    const B = Math.floor(Math.random()*99)
    let color = "#"+R+""+G+""+B
    while (color.length < 7) color += "F"
    return color
}

const reco = bidule.toRecords()

await display(
  Plot.plot({
    color: {legend: true},
    title: "Atih / region",
    width: 1300,
    height: 530,
    marginBottom: 50,
    x: { domain: reco.dep_name, tickRotate: -90 },
    y: { domain: reco.tauxPatients },
    marks: [
            Plot.barY(reco, {x: "dep_name", y: "tauxPatients", fill: "green"})
    ],
    document,
  })
);
```
